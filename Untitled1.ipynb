```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("netflix_titles.csv.zip")
```

```python
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df = df.dropna(subset=['date_added'])
```

```python
type_count = df['type'].value_counts()

plt.figure(figsize=(6,4))
type_count.plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

```python
df['date_added'] = pd.to_datetime(
    df['date_added'],
    errors='coerce'
)
```

```python
print(df['date_added'].isna().sum())
```

```python
df = df.dropna(subset=['date_added'])
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['date_added'].head(20)
```

```python
print(df['date_added'].dtype)
```

```python
plt.figure(figsize=(8,5))
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title("Ratings Distribution")
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10,5))
top_genres.plot(kind='bar')
plt.title("Top Genres")
plt.show()
```

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min', '')
movies['duration'] = pd.to_numeric(movies['duration'])
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = (
    movies['duration']
    .str.replace(' min', '', regex=False)
    .astype(int)
)
```

```python
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_directors.plot(kind='bar')
plt.title("Top Directors")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))

df['type'].value_counts().plot(
    kind='bar',
    ax=ax[0],
    title="Movies vs TV Shows"
)

df['rating'].value_counts().head(5).plot(
    kind='bar',
    ax=ax[1],
    title="Top Ratings"
)

plt.show()
```

```python
content_growth = df.groupby('year_added').size()

plt.figure(figsize=(10,5))

content_growth.plot()

plt.title("Netflix Content Growth")
plt.xlabel("Year")
plt.ylabel("Titles Added")

plt.show()
```

```python
movies['duration'] = pd.to_numeric(
    movies['duration'],
    errors='coerce'
)

movies = movies.dropna(subset=['duration'])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(movies['duration'], bins=20)

plt.title("Movie Duration Analysis")
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")

plt.show()
```

```python
shows = df[df['type']=="TV Show"].copy()

shows['seasons'] = (
    shows['duration']
    .str.replace(' Seasons','', regex=False)
    .str.replace(' Season','', regex=False)
)

shows['seasons'] = pd.to_numeric(
    shows['seasons'],
    errors='coerce'
)

shows = shows.dropna(subset=['seasons'])

shows.sort_values(
    by='seasons',
    ascending=False
)[['title','seasons']].head(10)
```
